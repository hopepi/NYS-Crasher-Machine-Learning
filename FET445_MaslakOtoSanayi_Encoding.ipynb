```python
import pandas as pd
import numpy as np

df = pd.read_csv('NYC_Crash_Cleaned_Severity.csv', low_memory=False)

print("--- BAŞLANGIÇ DURUMU ---")
print(f"Veri Boyutu: {df.shape}")
print(df[['SEVERITY', 'BOROUGH']].head())

severity_map = {
    'Property Damage Only': 0,
    'Injury': 1,
    'Fatal': 1 # Daha sonradan 1 çevrildi 
}

df['SEVERITY'] = df['SEVERITY'].map(severity_map)

print("SEVERITY Sayısallaştırıldı (0, 1, 1)")
print(f"Güncel Boyut: {df.shape}")
print(df[['SEVERITY']].head())
print(df.columns)
print(df.shape)
df.head(50)
```

```python
df = pd.get_dummies(df, columns=['BOROUGH'], prefix='BOROUGH', dtype=int)

print("BOROUGH One-Hot İşlemi Yapıldı")

print("--- DÖNÜŞÜM SONRASI ---")

new_columns = [col for col in df.columns if 'BOROUGH' in col or 'SEVERITY' in col]
print(df[new_columns].head())

print(f"\nSon Boyut: {df.shape}")
print("Sütun Listesi:", df.columns.tolist())
print(df.columns)
print(df.shape)
df.head(50)
```

```python
def map_vehicle_type_precise(vehicle):
    v = str(vehicle).lower().strip()
    
    if v in ['nan', 'unknown', 'unspecified', 'other', '', 'unk']:
        return 'Unknown'
    

    elif any(k in v for k in ['ambul', 'fire', 'police', 'ems', 'fdny', 'nypd']):
        return 'Emergency'
    

    elif any(k in v for k in ['bike', 'bicycle', 'motorcycle', 'scooter', 'moped', 'minibike', 'pedicab', 'vespa']):
        return 'Two Wheeler'
    

    elif any(k in v for k in ['taxi', 'livery', 'cab', 'limo', 'black car', 'uber']):
        return 'Taxi/Commercial'
    

    elif any(k in v for k in ['bus', 'omnib', 'truck', 'tractor', 'tanker', 'dump', 'flat', 'concrete', 
                              'garbage', 'refuse', 'semi', 'trailer', 'mixer', 'tow', 'wrecker', 
                              'com veh', 'comm', 'freight', 'box', 'moving', 'armored', 'mail', 'usps']):
        return 'Heavy Duty'
    

    elif any(k in v for k in ['sedan', 'station', 'passenger', 'suv', 'van', 'wagon', 'pick-up', 
                              'pickup', 'pk', 'convertible', 'minivan', 'carry all', '4 dr', '2 dr', 'sport utility']):
        return 'Passenger Car'
    
    else:
        return 'Other'


print("Araç türleri gruplanıyor...")

df['VEHICLE_1_GROUP'] = df['VEHICLE TYPE CODE 1'].apply(map_vehicle_type_precise)
df['VEHICLE_2_GROUP'] = df['VEHICLE TYPE CODE 2'].apply(map_vehicle_type_precise)

df['IS_MULTI_VEHICLE'] = np.where(df['VEHICLE_2_GROUP'] == 'Unknown', 0, 1)

print(" Yeni Sütunlar Eklendi: VEHICLE_1_GROUP, VEHICLE_2_GROUP, IS_MULTI_VEHICLE")


cols_to_drop = ['VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2']
df.drop(columns=cols_to_drop, inplace=True)
print(" Eski 'VEHICLE TYPE CODE' sütunları silindi.")


print("--- YENİ ARAÇ GRUPLARI (İLK 5 SATIR) ---")
print(df[['VEHICLE_1_GROUP', 'VEHICLE_2_GROUP', 'IS_MULTI_VEHICLE']].head())

print(f"Güncel Boyut: {df.shape}")
print("Sütun Listesi:", df.columns.tolist())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", context="talk")
plt.rcParams['figure.figsize'] = (16, 8)


target_cols = [
    'VEHICLE_1_GROUP', 
    'VEHICLE_2_GROUP',
    'CONTRIBUTING FACTOR VEHICLE 1',
    'CONTRIBUTING FACTOR VEHICLE 2'
]

for col in target_cols:
    print("\n" + "="*60)
    print(f" ANALİZ: {col}")
    print("="*60)
    
    unique_count = df[col].nunique()
    print(f"Toplam {unique_count} farklı kategori var.")
    
    top_20 = df[col].value_counts().head(20)
    total_rows = len(df)
    
    print("En Çok Görülen İlk 20 Değer ---")
    stats_df = pd.DataFrame({
        'Kategori': top_20.index,
        'Sayı': top_20.values,
        'Oran (%)': ((top_20.values / total_rows) * 100).round(2)
    })
    print(stats_df)
    
    temp_col = df[col].astype(str).str.lower()
    unknown_count = temp_col.isin(['nan', 'unknown', 'unspecified', 'other', 'unk']).sum()
    unknown_pct = (unknown_count / total_rows) * 100
    
    print(f"\n BİLİNMEYEN (Unknown/Unspecified/Other) TOPLAMI: {unknown_count:,} (%{unknown_pct:.2f})")
    
    plt.figure(figsize=(16, 6))
    sns.barplot(x=stats_df['Sayı'], y=stats_df['Kategori'], palette='viridis')
    plt.title(f'{col} - İlk 20 Dağılımı', fontsize=16, fontweight='bold')
    plt.xlabel('Kayıt Sayısı')
    plt.ylabel('')
    plt.show()

print("="*60)
print("ANALİZ TAMAMLANDI")
```

```python
def map_factor_precise(factor):
    f = str(factor).lower().strip()
    
    if f in ['nan', 'unknown', 'unspecified', 'other', '', 'unk', '1']:
        return 'Unknown'
    
    elif any(k in f for k in ['attention', 'distraction', 'cell phone', 'eating', 'texting', 'passenger', 'outside car']):
        return 'Distracted'
    
    elif any(k in f for k in ['speed', 'aggressive', 'following too closely', 'passing', 'lane', 
                              'yield', 'traffic control', 'turning', 'backing', 'tailgating', 'disregarded']):
        return 'Aggressive Driving'
    
    elif any(k in f for k in ['alcohol', 'drugs', 'medication', 'prescription', 'illness', 'consciousness', 'physical disability']):
        return 'Impaired'
    
    elif any(k in f for k in ['fatigued', 'asleep', 'drowsy']):
        return 'Fatigued'
    
    elif any(k in f for k in ['brakes', 'tire', 'steering', 'accelerator', 'failure', 'defective', 'tow hitch', 'oversized']):
        return 'Mechanical Failure'
    
    elif any(k in f for k in ['pavement', 'slippery', 'glare', 'view', 'obstruction', 'debris', 'animal', 'weather']):
        return 'Environmental'
    
    else:
        return 'Other Factor'


print("Kaza nedenleri gruplanıyor")

df['FACTOR_1_GROUP'] = df['CONTRIBUTING FACTOR VEHICLE 1'].apply(map_factor_precise)
df['FACTOR_2_GROUP'] = df['CONTRIBUTING FACTOR VEHICLE 2'].apply(map_factor_precise)


cols_to_drop_factors = ['CONTRIBUTING FACTOR VEHICLE 1', 'CONTRIBUTING FACTOR VEHICLE 2']
df.drop(columns=cols_to_drop_factors, inplace=True)

print(" Eski Faktör sütunları silindi.")
print(" Yeni Sütunlar: FACTOR_1_GROUP, FACTOR_2_GROUP")


print("--- 1. KAZA NEDENLERİ DAĞILIMI ---")
print(df['FACTOR_1_GROUP'].value_counts())
print("--- 2. CRASH CAUSES DISTRIBUTION ---")
print(df['FACTOR_2_GROUP'].value_counts())
df.head(50)
```

```python
import pandas as pd

print(f"İşlem Öncesi Boyut: {df.shape}")


print("Faktörlerdeki 'Bilinmeyen' veriler anlamlandırılıyor...")

if 'FACTOR_2_GROUP' in df.columns:
    df['FACTOR_2_GROUP'] = df['FACTOR_2_GROUP'].replace('Unknown', 'No Contribution')
    print(" Faktör 2: Unknown -> No Contribution işlemi yapıldı.")

if 'FACTOR_1_GROUP' in df.columns:
    df['FACTOR_1_GROUP'] = df['FACTOR_1_GROUP'].replace('Unknown', 'Unspecified')
    print(" Faktör 1: Unknown -> Unspecified işlemi yapıldı.")


cols_to_encode = ['FACTOR_1_GROUP', 'FACTOR_2_GROUP']

cols_to_encode = [col for col in cols_to_encode if col in df.columns]

if cols_to_encode:
    print(f"\nBu sütunlar One-Hot (Sayısal) yapılıyor: {cols_to_encode}")
    df = pd.get_dummies(df, columns=cols_to_encode, dtype=int)
else:
    print("\n Faktör sütunları bulunamadı (Zaten kodlanmış olabilir).")

print("-" * 40)
print(" İŞLEM TAMAMLANDI! Sadece Faktörler işlendi.")
print(f"Son Sütun Sayısı: {df.shape[1]}")
print("-" * 40)

print(df.head())
print("\nSütun Listesi:")
print(df.columns.tolist())
```

```python
import pandas as pd
import numpy as np
import re

df = pd.read_csv('NYC_Crash_ML_Final.csv', low_memory=False)
print(f"Veri Yüklendi: {df.shape}")


def get_road_category(name):
    if pd.isna(name) or str(name).strip() == '' or 'UNKNOWN' in str(name).upper():
        return 'UNKNOWN_OTHER'

    name = str(name).upper().strip()
    
    
    if any(x in name for x in ['EXPRESSWAY', 'PARKWAY', 'HIGHWAY', 'THRUWAY', 'CONCOURSE', 'BOULEVARD', 'TURNPIKE', 'BLVD', 'PKWY', 'EXPY', 'HWY']):
        return 'ARTERIES_HIGHWAYS'
    
    elif any(x in name for x in ['AVENUE', 'STREET', 'ROAD', 'AVE', 'ST', 'RD']):
        return 'LOCAL_MAIN_ROADS'
    
    elif any(x in name for x in ['PLACE', 'PLAZA', 'CIRCLE', 'COURT', 'LOOP', 'LANE', 'TERRACE', 'PL', 'CIR']):
        return 'SPECIAL_AREAS'
    
    else:
        return 'UNKNOWN_OTHER'

print("Sokak isimleri analiz ediliyor ve 3 ana gruba ayrılıyor...")
df['ROAD_TYPE'] = df['CROSS STREET NAME'].apply(get_road_category)


print("Gruplar 0 ve 1 sütunlarına dönüştürülüyor...")


df = pd.get_dummies(df, columns=['ROAD_TYPE'], prefix='ROAD_TYPE', dtype=int)


df.drop(columns=['CROSS STREET NAME'], inplace=True)


if 'ROAD_TYPE_UNKNOWN_OTHER' in df.columns:
    df.drop(columns=['ROAD_TYPE_UNKNOWN_OTHER'], inplace=True)
    print(" 'ROAD_TYPE_UNKNOWN_OTHER' sütunu silindi")


print("-" * 40)
print(f" İŞLEM TAMAMLANDI! Son Boyut: {df.shape}")
print("-" * 40)

new_cols = [col for col in df.columns if 'ROAD_TYPE' in col]
print("Yeni Eklenen Sütunlar:")
print(new_cols)

print("\nÖrnek Veri:")
print(df[new_cols].head())
```
